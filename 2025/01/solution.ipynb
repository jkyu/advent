```python
with open("input.txt", "r") as f:
    lines = f.readlines()
lines = [line.strip() for line in lines]
```

```python
# part 1
position = 50
password = 0
for line in lines:
    direction = line[0]
    distance = int(line[1:])
    if direction == "L":
        position = (position - distance) % 100
    else:
        position = (position + distance) % 100
    password += (position == 0)
print(password)
```

```python
# part 2
position = 50
password = 0
for line in lines:
    direction = line[0]
    distance = int(line[1:])
    if direction == "L":
        if (position <= distance):
            password += (position != 0) + (distance - position) // 100
        position = (position + 100 - distance) % 100
    else:
        password += (position + distance) // 100
        position = (position + distance) % 100
print(password)
```
